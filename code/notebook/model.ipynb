# Feature Types

```python
cat_cols=x.select_dtypes(include=['object','category']).columns.to_list()
num_cols=x.select_dtypes(include=[np.number]).columns.to_list()
```

```python
print("Categorical Columns are : ",cat_cols)
print("numerical Columns are : ",num_cols)
```

# preprocessing pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
numerical_transformer=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',StandardScaler())
])
```

```python
categorical_transformer=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('num',numerical_transformer,num_cols),
    ('cat',categorical_transformer,cat_cols)
],    remainder='drop')
```

```python
preprocessor
```

# Stage - 1 Classification baseline

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
y=(y=='approved').astype(int)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y)
print("Train size :",x_train.shape)
print("Test Size : ",x_test.shape)
```

```python
rf_clf_pipeline=Pipeline(steps=[
('preprocessor',preprocessor),
('rf_cls',RandomForestClassifier())
]
)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    'rf_cls__n_estimators':[100,200,300,400],
    'rf_cls__max_depth':[None,4,8,10],
    'rf_cls__max_features':['sqrt',None]
}
grid=GridSearchCV(rf_clf_pipeline,param_grid,n_jobs=-1,scoring='f1',cv=5)
grid.fit(x_train,y_train)
print("best params are : ",grid.best_params_)
best_clf=grid.best_estimator_
y_pred_grid=best_clf.predict(x_test)
print("Classification Report is : ",classification_report(y_test,y_pred_grid))
```

# Fit RF baseline pipelines

```python
rf_clf_pipeline=Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('rf_cls',RandomForestClassifier(n_estimators=400,max_depth=None,max_features=None,random_state=10,oob_score=True))
])
rf_clf_pipeline.fit(x_train,y_train)
y_pred_cls=rf_clf_pipeline.predict(x_test)
print("Classification Report is :\n",classification_report(y_test,y_pred_cls))
print("OOB score is :",getattr(rf_clf_pipeline.named_steps['rf_cls'],'oob_score_',None))
```

```python
rf_clf_pipeline
```

# checking feature importance

```python
feature_importance=rf_clf_pipeline.named_steps['rf_cls'].feature_importances_

preprocessor=rf_clf_pipeline.named_steps['preprocessor']

feature_names=preprocessor.get_feature_names_out()

importance_df=pd.DataFrame(
    {
    'Features':feature_names,
    'Importance':feature_importance
    }
)
importance_df=importance_df.sort_values(by='Importance',ascending=False)


print("Top 10 Importance Features :\n",importance_df.head(10))

plt.figure(figsize=(10,6))
plt.barh(importance_df['Features'][:10],importance_df['Importance'][:10])
plt.xlabel("Random Forest Features Importance")
plt.ylabel("Features")
plt.title("TOP 10 Importance Features ")
plt.show()
```
